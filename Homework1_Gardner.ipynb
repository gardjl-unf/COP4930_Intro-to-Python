```python
# importing needed packages
import matplotlib.pyplot as plt
import numpy as np
import heapq
import time
import matplotlib.animation as animation

# Point class to represent a 2D point with x and y coordinates
class Point:
    # constructor
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

```python
# Three functions to draw board, grid on the board, and lines
def draw_board():
    # create a figure to draw the board
    fig = plt.figure(figsize=[10,10])
    ax = fig.add_subplot(111)
    return fig, ax

def draw_grid(ax):
    # draw the vertical lines
    for x in range(50):
        ax.plot([x, x], [0,50-1], color = '0.75', linestyle='dotted')
    # draw the horizontal lines
    for y in range(50):
        ax.plot([0, 50-1], [y,y], color = '0.75', linestyle='dotted')
    ax.set_position([0,0.02,1,1])

def draw_line(ax, xs, ys):
    ax.plot(xs, ys, color='k')
```

```python
# call draw_board function to draw board
fig, ax = draw_board()
# call draw_grid function to draw grid on the board
draw_grid(ax)

# Draw a letter with points that are offset x,y from the previous
def draw_letter(p1, offsets):
	p2 = Point(0,0)
	for o in offsets:
		p2 = Point(p1.x + o.x, p1.y + o.y)
		draw_line(ax, [p1.x,p2.x], [p1.y,p2.y])
		p1 = p2

# Draw P
p1 = Point(2.5,30)
offsets = [Point(4,0),Point(1,-1),Point(0,-3),Point(-1,-1),Point(-3,0),Point(0,-5),Point(-1,0), Point(0,10)]
draw_letter(p1,offsets)

p1 = Point(3.5,29)
offsets = [Point(2,0),Point(1,-1),Point(0,-1),Point(-1,-1),Point(-2,0),Point(0,3)]
draw_letter(p1,offsets)

# Draw Y
p1 = Point(7.5,30)
offsets = [Point(1,0),Point(1.5,-3.5),Point(1.5,3.5),Point(1,0),Point(-2,-4.5),Point(0,-6), Point(-1,0),Point(0,6),Point(-2,4.5)]
draw_letter(p1,offsets)

# Draw T
p1 = Point(13,30)
offsets = [Point(5,0),Point(0,-1),Point(-2,0),Point(0,-9),Point(-1,0),Point(0,9), Point(-2,0),Point(0,1)]
draw_letter(p1,offsets)

# Draw H
p1 = Point(18.5,30)
offsets = [Point(1,0),Point(0,-4.5),Point(3,0),Point(0,4.5),Point(1,0),Point(0,-10), Point(-1,0),Point(0,4.5),Point(-3,0),Point(0,-4.5),Point(-1,0),Point(0,10)]
draw_letter(p1,offsets)

# Draw O
p1 = Point(25,30)
offsets = [Point(3,0),Point(1,-1),Point(0,-8),Point(-1,-1),Point(-3,0), Point(-1,1),Point(0,8),Point(1,1)]
draw_letter(p1,offsets)

p1 = Point(25,28)
offsets = [Point(1,1),Point(1,0),Point(1,-1),Point(0,-6),Point(-1,-1),Point(-1,0),Point(-1,1),Point(0,6)]
draw_letter(p1,offsets)

# Draw N
p1 = Point(29.5,30)
offsets = [Point(1.5,0),Point(3,-9),Point(0,9),Point(1,0),Point(0,-10), Point(-1.5,0),Point(-3,9),Point(0,-9),Point(-1,0),Point(0,10)]
draw_letter(p1,offsets)

# Draw J
p1 = Point(36,14)
offsets = [Point(1,0),Point(0,-3),Point(1,-1),Point(1,0),Point(1,1), Point(0,8),Point(1,0),Point(0,-9),Point(-1,-1),Point(-3,0), Point(-1,1),Point(0,4)]
draw_letter(p1,offsets)

# Draw G
p1 = Point(41.5,18)
offsets = [Point(1,1),Point(3,0),Point(1,-1),Point(0,-2),Point(-1,0),Point(0,1),Point(-1,1),Point(-1,0),Point(-1,-1),Point(0,-6),Point(1,-1),Point(1,0),Point(1,1),Point(0,2),Point(-2,0),Point(0,1),Point(3,0),Point(0,-4),Point(-1,-1),Point(-3,0),Point(-1,1),Point(0,8)]
draw_letter(p1,offsets)

# show plot
plt.show()
```
