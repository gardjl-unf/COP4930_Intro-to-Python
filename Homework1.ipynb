```python
# importing needed packages
import matplotlib.pyplot as plt
import numpy as np
import heapq
import time
import matplotlib.animation as animation

# Point class to represent a 2D point with x and y coordinates
class Point:
    # constructor
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

```python
# Three functions to draw board, grid on the board, and lines
def draw_board():
    # create a figure to draw the board
    fig = plt.figure(figsize=[10,10])
    ax = fig.add_subplot(111)
    return fig, ax

def draw_grid(ax):
    # draw the vertical lines
    for x in range(50):
        ax.plot([x, x], [0,50-1], color = '0.75', linestyle='dotted')
    # draw the horizontal lines
    for y in range(50):
        ax.plot([0, 50-1], [y,y], color = '0.75', linestyle='dotted')
    ax.set_position([0,0.02,1,1])

def draw_line(ax, xs, ys):
    ax.plot(xs, ys, color='k')
```

```python
# call draw_board function to draw board
fig, ax = draw_board()
# call draw_grid function to draw grid on the board
draw_grid(ax)

# draw letter X
p1 = Point(10,30)
p2 = Point(20,20)
draw_line(ax, [p1.x,p2.x], [p1.y,p2.y])
p3 = Point(20,30)
p4 = Point(10,20)
draw_line(ax, [p3.x,p4.x], [p3.y,p4.y])

# draw letter Y
p1 = Point(30,30)
p2 = Point(34,26)
p3 = Point(38,30)
p4 = Point(34,20)
draw_line(ax, [p1.x,p2.x], [p1.y,p2.y])
draw_line(ax, [p2.x,p3.x], [p2.y,p3.y])
draw_line(ax, [p2.x,p4.x], [p2.y,p4.y])

# show plot
plt.show()
```
